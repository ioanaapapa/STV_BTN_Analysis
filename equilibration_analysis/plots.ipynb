```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import MDAnalysis as mda
```

```python
def get_df(file,columns):
    datas = open(file,'r').readlines() #insert the path here

    data = []

    for line in datas[28:]:
        line = line.split()
        data.append([float(a) for a in line])

    df = pd.DataFrame(data)
    df.columns=columns
    return df
```

```python
def plot(df1,df2,df3,df4,df5, column):
    time = df1["Time(ps)"]
    fig = plt.figure()
    fig, ax = plt.subplots()
    ax.plot(time, df1[column], label='1')
    ax.plot(time, df2[column], label='2')
    ax.plot(time, df3[column], label='3')
    ax.plot(time, df4[column], label='4')
    ax.plot(time, df5[column], label='5')
    ax.set_xlabel('Time(ps)')
    ax.set_ylabel(column)
    ax.legend()
    plt.grid()
    plt.show()
    plt.close()
    
```

### Entropy calculation - with previous CodeEntropy release

```python
tprfile1 = "rep_1/prod_BTN_only.tpr"
trrfile1 = "rep_1/prod_BTN_only.trr"
u1 = mda.Universe(tprfile1, trrfile1)
```

```python
from CodeEntropy.ClassCollection import DataContainer as DC
dataContainer = DC.DataContainer (u1)
```

```python
num_frames = len(dataContainer.trajSnapshots)
print(num_frames)
```

```python
dataContainer
```

```python
from CodeEntropy.FunctionCollection import EntropyFunctions as EF
selection_string = "all"
axis_list = ["N2","C4","C2"]
outFile = None
moutFile = None
nmdFile = None
csv_out = None
tScale = 1.0
fScale = 1.0
temper = 298.0
verbose = 3 
thread = 4
```

```python
wm_entropyFF1, wm_entropyTT1 = EF.compute_entropy_whole_molecule_level (
    arg_hostDataContainer = dataContainer,
    arg_outFile = outFile,
    arg_selector = selection_string,
    arg_moutFile = moutFile,
    arg_nmdFile = nmdFile,
    arg_fScale = fScale,
    arg_tScale = tScale,
    arg_temper = temper,
    arg_verbose = verbose
)
print (f"wm_entropyFF = {wm_entropyFF1}")
print(f"wm_entropyTT = {wm_entropyTT1}")
```

```python
UA_entropyFF1, UA_entropyTT1, res_df = EF.compute_entropy_UA_level (
    arg_hostDataContainer = dataContainer,
    arg_outFile = outFile,
    arg_selector = selection_string,
    arg_moutFile = moutFile,
    arg_nmdFile = nmdFile,
    arg_fScale = fScale,
    arg_tScale = tScale,
    arg_temper = temper,
    arg_verbose = verbose,
    arg_axis_list = axis_list,
    arg_csv_out = csv_out,
)
print (f"UA_res_entropyFF = {UA_entropyFF1}")
print(f"UA_entropyTT = {UA_entropyTT1}")    
```

```python
result_entropyAEM1 = EF.compute_topographical_entropy_AEM(
    arg_hostDataContainer = dataContainer,
    arg_selector = selection_string,
    arg_outFile = outFile,
    arg_verbose = verbose
)
print(f"result_entropyAEM = {result_entropyAEM1}")
```

```python
total1 = wm_entropyFF1 + wm_entropyTT1 + UA_entropyFF1 + UA_entropyTT1 + result_entropyAEM1
print(f"Total Entropy = {total1} J/mol/K")
```
